```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

#### Load Image

```python
image = mpl.image.imread("flower.jpg")
```

```python
plt.imshow(image)
```

##### Flatten each pixel RGB value into a single row

```python
image.shape
```

```python
X = image.reshape(-1,3)
X.shape
```

##### Perform k-means clsutering

```python
kmeans = KMeans(n_clusters=2,n_init=10)
```

```python
kmeans.fit(X)
```

```python
segmented_img = kmeans.cluster_centers_[kmeans.labels_]
segmented_img = segmented_img.reshape(image.shape)# reshape back to org shape of image
```

```python
plt.imshow(segmented_img/255)
```

```python
cluster_centers= kmeans.cluster_centers_
cluster_centers
```

```python
cluster_names = ["Cluster 1", "Cluster 2"]

# Print out the main colors for each cluster
for i, center in enumerate(cluster_centers):
    print(f"The main color of {cluster_names[i]} is: RGB({int(center[0])}, {int(center[1])}, {int(center[2])})")
```

##### As we increase the number of clusters (n_clusters), the image gets more detailed because more colors are used to represent different segments of the image. This results clearer details.


```python
kmeans1 = KMeans(n_clusters=10,n_init=10)
```

```python
kmeans1.fit(X)
```

```python
segmented_img = kmeans1.cluster_centers_[kmeans1.labels_]
segmented_img = segmented_img.reshape(image.shape)# reshape back to org shape of image
```

```python
plt.imshow(segmented_img/255)
```
